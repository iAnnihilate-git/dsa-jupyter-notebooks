```python
from pymysql import *
from tkinter import *
from tkinter import messagebox as msg
class login(Frame):
    def init(self,master):
        super().init(master)
        self.l1=Label(self,text='User Name',fg='blue',bg='light gray',
        font=('algerian',13))
        self.l2=Label(self,text='Password',fg='blue',bg='light gray',
        font=('algerian',13))
        
        self.t1=Entry(self,fg='blue',bg='light gray',
            font=('arial',14),justify='center',bd=4)
        self.t2=Entry(self,show='#',fg='blue',bg='light gray',
            font=('arial',14),justify='center',bd=4)
        
        self.b1=Button(self,text='Login',bg='light blue',fg='red',
        font=('algerian',15),bd=4,command=self.check)
        
        self.b2=Button(self,text='Add',bg='light blue',fg='red',
        font=('algerian',15),bd=4,command=self.save)
        
        self.rowconfigure(index=0,pad=15)
        self.rowconfigure(index=1,pad=15)
        self.rowconfigure(index=2,pad=15)
        
        self.columnconfigure(index=0,pad=15)
        self.columnconfigure(index=1,pad=15)
        
        self.l1.grid(row=0,column=0)
        self.t1.grid(row=0,column=1)
        
        self.l2.grid(row=1,column=0)
        self.t2.grid(row=1,column=1)
        
        self.b1.grid(columnspan=2)
        self.b2.grid(columnspan=2)
        self.pack()
    def check(self):
        con=connect(db='ducat',user='root',passwd='admin',host='localhost')
        cur=con.cursor()
        uid=self.t1.get()
        pwd=self.t2.get()
        cur.execute("select * from login where userid='%s' and passwd='%s'"%(uid,pwd))
        rs=cur.fetchall()
        if len(rs)>=1:
            msg.showinfo('logininfo','Thanks for login')
        else:
            msg.showerror('loginerror','Sorry')
        con.close()
    def save(self):
        con=connect(db='ducat',user='root',passwd='admin',host='localhost')
        cur=con.cursor()
        uid=self.t1.get()
        pwd=self.t2.get()
        i=cur.execute("update login set passwd='%s' where userid='%s'"%(pwd,uid))
        if i==1:
            print('Updated')
            con.commit()
            self.t1.delete(0,'end')
            self.t2.delete(0,'end')
        con.close()
root=Tk()
obj=login(root)
root.title('login')
root.geometry('450x200')
root.mainloop()
```
